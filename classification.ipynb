```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
w1=pd.read_csv("wisc_bc_data.csv")
```

```python
w1.head()
```

```python
w1.isnull().sum()
w1.info()
```

```python
w1.describe().T
```

```python
for i in w1.columns:
    if w1[i].dtype in ["int64","float64"]:
        w1[i].replace(to_replace=0,value=w1[i].median(),inplace=True)
w1.head(45)      
w1["diagnosis"]=w1["diagnosis"].replace(to_replace='M',value=0)
w1["diagnosis"]=w1["diagnosis"].replace(to_replace='B',value=1)
w1.head()
```

```python
w_encoded=pd.get_dummies(w1)
```

```python
w_encoded.columns
w_encoded.head()
```

```python
round(w1.var())
w=w1
w.var()
```

```python
w.corr()
```

```python
from sklearn.model_selection import train_test_split
```

```python
w_train,w_test=train_test_split(w,test_size=0.25,random_state=100)
w_train.shape,w_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
feature=w.drop(columns=["diagnosis"])
features=['perimeter_mean',
       'area_mean', 'area_se', 'perimeter_worst',
       'area_worst']
#features=["id","area_mean","area_se","perimeter_worst","area_worst"]
```

```python
target=["diagnosis"]
w.columns
```

```python
model.fit(w_train[features],w_train[target])
```

```python
train_accuracy=model.score(w_train[features],w_train[target])
test_accuracy=model.score(w_test[features],w_test[target])
train_accuracy,test_accuracy
```

K-NN

```python
k=w
k.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
k1=k.drop(columns="diagnosis")
```

```python
#s_values=scaler.fit_transform(k[["age","TB","DB","alkphos","sgpt","sgot","TP","ALB","A_G"]])
#k["age"]=s_values[:,0]
s_values=scaler.fit_transform(k1)
c=0
for i in k1.columns:
    k1[i]=s_values[:,c]
    c+=1
k1.head()
k1["diagnosis"]=k["diagnosis"]
k1.var()
```

```python
from sklearn.model_selection import train_test_split
w2_train,w2_test=train_test_split(k1,test_size=0.15,random_state=100)
w2_train.shape,w2_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model3=KNeighborsClassifier(n_neighbors=3,metric="euclidean")
```

```python
#features=["perimeter_worst","texture_worst"]
#features=['perimeter_mean','area_mean', 'area_se', 'perimeter_worst','area_worst']
features=['radius_mean', 'texture_mean', 'perimeter_mean','points_worst']
      # 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       #'points_mean', 'symmetry_mean', 'dimension_mean', 'radius_se',
       #'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       #'compactness_se', 'concavity_se', 'points_se', 'symmetry_se',
       #'dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst',
       #'area_worst']
target=['diagnosis']
```

```python
model3.fit(w2_train[features],w2_train[target])
```

```python
train_accuracy2=model3.score(w2_train[features],w2_train[target])
test_accuracy2=model3.score(w2_test[features],w2_test[target])
train_accuracy2,test_accuracy2
```
