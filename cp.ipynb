```python
b=b.drop(columns=["capital-gain","capital-loss"])
```

```python
q1=b.quantile(0.25)
q2=b.quantile(0.50)
q3=b.quantile(0.75)
iqr=q3-q1
r1=iqr*1.5+q3
r2=q1-(iqr*1.5)

r1
```

```python
r2
```

```python
c=['age','fnlwgt','education-num','hours-per-week']
```

```python
for i,j,k in zip(r1,c,r2):
    b=b[b[j]<=i]
    b=b[b[j]>=k]
b.shape
b
```

```python
b['h']=pd.cut(b['hours-per-week'],bins=[30,40,50,60],labels=[40,50,60])
```
